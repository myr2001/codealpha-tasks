```python
'''
Import modules that will be used in the project
'''
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

'''
Pandas allows us to read csv files and place them into a dataframe.
You may get your datasets in csv format or other formats from Kaggle.
'''
df = pd.read_csv("Iris.csv")
print(df.info())

'''
Show dataframe
'''
df
```

```python
'''
Scatter plot
'''
df.plot.scatter(x="SepalWidthCm", y="SepalLengthCm", color="r") # scatter plot
plt.ylabel("Sepal Length")
plt.xlabel("Sepal Width")
plt.title("Scatter Plot of Sepal Width & Length")
# plt.grid()
```

```python

'''
Pie Chart
'''
values=df.Species.value_counts()
labels=values.index
plt.figure(figsize=(10,10))
plt.pie(values, labels=labels, explode=[0.1,0,0]) # explode separates a slice
```

```python

'''
Histogram
'''
mu = round(df["PetalWidthCm"].mean(),2)
sigma = round(df["PetalWidthCm"].std(),2)
num_bins=40
# semicolon to avoid the texts and just get the graph
plt.hist(df["PetalWidthCm"], num_bins, facecolor="blue");
plt.title(f"Histogram: mu = {mu}, std = {sigma}");
```

```python
'''
Multiple Plots
'''

plt.figure(figsize=(20,10))

plt.subplot(1,2,1) # row, column, plotting 1st one rn
plt.plot(df["PetalWidthCm"], "r--") # r-- means red color and dashed wline
plt.grid()

plt.subplot(1,2,2) # row, column, plotting 2nd one rn
plt.plot(df["PetalLengthCm"],"b.") # b. means blue color and dots
```

```python
'''
Multiple Plots
'''

plt.figure(figsize=(20,10))

plt.subplot(2,1,1) # row, column, plotting 1st one rn
plt.plot(df["PetalWidthCm"], "r--") # r-- means red color and dashed wline
plt.grid()

plt.subplot(2,1,2) # row, column, plotting 2nd one rn
plt.plot(df["PetalLengthCm"],"b.") # b. means blue color and dots
```

```python
'''
3D Plots
'''

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (15, 15))
ax = fig.add_subplot(111, projection = '3d') # 111: row, column, subplot

x = df.PetalWidthCm
y = df.PetalLengthCm
z = df.SepalWidthCm

colors = np.random.rand(len(df.PetalWidthCm))

ax.scatter(x,y,z, c = colors, cmap="viridis", s=50)
ax.set_xlabel("Petal Width")
ax.set_ylabel("Petal Length")
ax.set_zlabel("Sepal Width")
```

```python
'''
Seaborn Scatterplot
'''
# Seaborn is preferred for big data
plt.figure(figsize=(5,5))
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", hue="Species", data=df)
```

```python
'''
Seaborn Count Plot
'''

plt.figure(figsize=(5,5))
sns.countplot(x="Species",data=df)
```

```python
'''
Seaborn Pairplot
'''

sns.pairplot(df, hue="Species",
             vars = ["SepalLengthCm","SepalWidthCm",
                     "PetalLengthCm", "PetalWidthCm"])
```

```python
'''
Seaborn Distplot
'''

sns.distplot(df[df["Species"]=="Iris-setosa"]["SepalLengthCm"], bins=25, color="b")
sns.distplot(df[df["Species"]=="Iris-versicolor"]["SepalLengthCm"], bins=25, color="g")
sns.distplot(df[df["Species"]=="Iris-virginica"]["SepalLengthCm"], bins=25, color="r")
```
